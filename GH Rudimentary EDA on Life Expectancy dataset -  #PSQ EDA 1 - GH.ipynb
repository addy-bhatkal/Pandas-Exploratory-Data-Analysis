```python
import os
os.chdir('C:\\Users\\ABC\\6. Datasets')
df = pd.read_csv('Life Expectancy Data 2 - Manipulated for Ridge.csv')
df.head()
import matplotlib 
import seaborn as sns
```

```python
df.columns
df1 = pd.DataFrame(df.groupby('Country')['Life expectancy '].mean())
df1['Std Dev'] = df.groupby('Country')['Life expectancy '].std()
df1['Std Dev'].mode().tail(50)
df1[df1['Life expectancy '].isna()] #IMP
mean = df['Life expectancy '].mean()
std = df['Life expectancy '].std()
df2 = df1['Life expectancy '].fillna(mean) 
df2 = df1.fillna(mean) 
df2['Life expectancy '].fillna(mean)
df2['Std Dev'].fillna(std)
```

```python
sns.kdeplot(df2['Std Dev'])
listys = []
listyb = []
for i in df2['Std Dev']:
    if i > 20: 
        listys.append(i)
    else:
        listyb.append(i)
a = sum(listys)/len(listys)
b = sum(listyb)/len(listyb)
print(a,b)
```
